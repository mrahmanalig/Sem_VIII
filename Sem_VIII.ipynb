**Q.2: Write a Python program to find the largest of three numbers using conditional statements.**

```python
# Problem 2:
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
n3 = int(input("Enter third number: "))

if (n1 > n2) and (n2 > n3):
  print("First is the greatest: ", n1)

elif(n2 > n1) and (n2 > n3):
  print("Second is the greatest: ", n2)

elif  (n1 == n2) and (n2 == n3):
  print("All are equal!!")

else:
    print("Third is the greatest: ", n3)
```

**Q.4: Write a Python program to count the number of vowels in a given string.**

```python
# Problem 4:
string = input("Enter any string: ")
count = 0
vowel = set("aeiouAEIOU")
for c in string:
  if c in vowel:
    count += 1
print("Number of vowels: ", count)
```

**Q.1: Create a list of numbers and find the sum and average of the list.**

```python
# Problem 1

n = int(input("Enter how many numbers: "))

numbers = []

for i in range(n):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

total = sum(numbers)
avg = total/n

print("List of numbers:", numbers)
print("Sum:", total)
print("Average is: ", avg)
```

**Q2: Write a function to check whether a given string is a palindrome.**

```python
# Problem 2
text = input("Enter a string: ")

reverse = ""
for char in text:
    reverse = char + reverse

if text == reverse:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

**Q4: Write a program to sort a list in ascending and descending order.**

```python
# Problem 4
numbers = [45, 12, 89, 23, 7,67,43,13]

asc = numbers.copy()
n = len(asc)

for i in range(n):
    for j in range(0, n - i - 1):
        if asc[j] > asc[j + 1]:
            temp = asc[j]
            asc[j] = asc[j + 1]
            asc[j + 1] = temp

print("Ascending order:", asc)

desc = numbers.copy()

for i in range(n):
    for j in range(0, n - i - 1):
        if desc[j] < desc[j + 1]:
            temp = desc[j]
            desc[j] = desc[j + 1]
            desc[j + 1] = temp

print("Descending order:", desc)
```

**Q3: Create a NumPy array from having numbers between 1 to 100 and print only the prime numbers in them.**

```python
# Problem 3
import numpy as np
nums = np.random.randint(1, 101, 30)

print("Random Numbers :",nums)

print("\nPrime Numbers: ")

for n in nums:
    if n > 1:
        for i in range(2, n):
            if n % i == 0:
                break
        else:
            print(n, end=" ")
```
